# 1. 单分支结构

```python
money = 10000  # 余额
s = int(input('请输入取款金额：'))  # 取款金额

# 判断余额是否充足
if s <= money :
    money -= s
    print('取款成功，余额为：', money)
```

# 2. 双分支结构

```python
money = 10000  # 余额
s = int(input('请输入取款金额：'))  # 取款金额

# 判断余额是否充足
if s <= money :
    money -= s
    print('取款成功，余额为：', money)
else:
    print('取款失败，余额不足！')
```

# 3. 多分支结构

```python
'''多分支结构，多选一执行
从键盘录入一个整数 成绩
90-100 A
80-89 B
70-79 C
60-69 D
0-59 E
小于0或者大于100 为非法数据（不是有效范围）'''
score = int(input('请输入一个成绩：'))
# 判断
if 90 <= score <= 100:
    print('A级')
elif 80 <= score <= 89:
    print('B级')
elif 70 <= score <= 79:
    print('C级')
elif 60 <= score <= 69:
    print('D级')
elif 0 <= score <= 59:
    print('E级')
else:
    print('数据非法')
```

# 4. 嵌套if使用

```python
'''会员  >=200 8折
        >100 9折
        <100 不打折
    非会员  >=200 95折
           <200 不打折'''
answer = input('你是会员吗？y/n：')
money = float(input('请输入您的购物金额：'))

# 外层判断是否是会员
if answer == 'y':
    print('是会员')
    if money >= 200:
        money *= 0.8
        print('最终金额：', money)
    elif 200 > money >= 100:
        money *= 0.9
        print('最终金额：', money)
    else:
        print('最终金额：', money)
else:  # 非会员
    print('不是会员')
    if money >= 200:
        money *= 0.95
        print('最终金额：', money)
    else:
        print('最终金额', money)
```
